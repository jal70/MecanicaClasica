Resolvemos la ecuación 
$$\frac{d}{dt}v+\frac{\beta}{m}v+g=0 $$

```python
#beta/m (1/s)
b=2
g=10
#condiciones iniciales, velocidad
tv0=0
v0=0
```

```python
#Función
y = function('y')(x)
f=diff(y,x,1) + b*y + g == 0
```

```python
#Resolución ecuación
h = desolve(f, y, ics=[tp0,v0])
```

```python
print "Unsimplified:"
print h
```

```python
print "Simplified:"
print h.expand()
```

```python
#Gráfica
plot( -h(x), 0, 2, gridlines="minor" )
```
