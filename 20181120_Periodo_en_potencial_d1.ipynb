```python
#Potencial
V = function('V')(x)
V = 20*x**2
```

```python
#Gráfica
plot( V(x), -100, 100, gridlines="minor" )
```

```python
#Puntos de retorno
E=30 #Referencia de energía mecánica
corte=E-V(x)==0 #Condición de cero energía cinética (punto de reposo)
```

```python
#Buscamos las raices
h=solve(corte,x)
```

```python
print "Lista de soluciones:"
print h
print "Primera raiz--> ",h[0]
print "Segunda raiz--> ",h[1]
print "Valor numérico de la primera raiz--> ",h[0].rhs()
print "Formato float de la primera raiz--> ",float(h[0].rhs())
```
