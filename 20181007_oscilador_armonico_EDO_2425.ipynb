Resolvemos la ecuación 
$$\frac{d^2}{dt^2}x+\omega^2x=0 $$

```python
#Frecuencia (1/s)
omega=2
#Condiciones iniciales, posición
tp0=0
x0=20
#condiciones iniciales, velocidad
tv0=0
v0=0
```

```python
#Función
y = function('y')(x)
f=diff(y,x,2) + omega**2*y == 0
```

```python
#Resolución ecuación
h = desolve(f, y, ics=[tp0,x0,v0])
```

```python
print "Unsimplified:"
print h
```

```python
print "Simplified:"
print h.expand()
```

```python
#Gráfica
plot( h(x), 0, 5, gridlines="minor" )
```
